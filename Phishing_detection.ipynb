<h1>5. Model training</h1>
<p>Supervised learning is a type of machine learning where an algorithm is trained to make predictions based on labeled data. In supervised learning, we use labeled data to teach the computer how to make predictions about new, unseen data. The algorithm learns from the labeled data by identifying patterns and relationships between the input features and the labeled outcomes. Once the algorithm has learned these patterns, it can make predictions about new data by applying these patterns to the new input features.</p>

There are two types of supervised learning - classification and regression.

Classification is a type of supervised learning where the goal is to predict which category or class an input example belongs to. In other words, we are trying to assign a label or category to each input example based on its features or attributes.

For example, in email spam detection, we might use classification to predict whether a new email is spam or not spam based on its subject, sender, and content. In medical diagnosis, we might use classification to predict whether a patient has a particular disease based on their symptoms and medical history.

Classification problems can be either binary or multi-class. Binary classification involves predicting one of two possible outcomes, such as "spam" or "not spam", while multi-class classification involves predicting one of more than two possible outcomes, such as "cat", "dog", or "bird".

Our data falls into the category of <b>binary classification</b>

<h1>Defining a helper function</h1>
<p>Evaluation of classification models is typically done using metrics such as accuracy, precision, recall, and F1 score. These metrics help to assess the performance of the algorithm and identify areas where it may need improvement.
</P>
<p>We will create a helper function where all of our result from different algorithms are going to be stored for future comparison.</p>

```python
model_name = []
accuracy_result  = []
f1_result = []
recall_result = []
precision_result = []

def algo_holder(model, acc, f1, recall, precision):
    if model not in model_name:
        model_name.append(model)
    accuracy_result.append(round(acc,5))
    f1_result.append(round(f1,5))
    recall_result.append(round(recall,5))
    precision_result.append(round(precision,5))
```

<h1>6.1 Logistic Regression</h1>
<p>Logistic regression is a type of machine learning algorithm used for classification tasks. It works by calculating the probability of a new input data point belonging to a certain class.

To understand how logistic regression works, imagine you want to predict if a student will pass or fail an exam based on their study hours.

The logistic regression algorithm starts by fitting a line to the data points that best separates the pass and fail students based on their study hours. This line is called the decision boundary.

Next, the algorithm calculates the probability of a new input student passing or failing the exam based on their study hours. It does this by plugging the student's study hours into the equation of the decision boundary line. The output of this equation is a value between 0 and 1, which represents the probability of the student passing the exam.

Finally, the algorithm assigns the class with the highest probability as the predicted class for the new input student. For example, if the probability of passing the exam is 0.8, the algorithm would predict that the student will pass.

In summary, logistic regression is like a probability calculator that calculates the probability of an input belonging to a certain class based on the decision boundary that best separates the classes. It is a simple but powerful algorithm for classification tasks.</p>

```python
from sklearn.linear_model import LogisticRegression

lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)
```

```python
y_train_lr = lr_model.predict(X_train)
y_test_lr = lr_model.predict(X_test)
```

```python
cm = metrics.confusion_matrix(y_train_lr, y_train)
cm_display = metrics.ConfusionMatrixDisplay(cm).plot()
```

```python
acc_test_lr = metrics.accuracy_score(y_test,y_test_lr)
f1_score_test_lr = metrics.f1_score(y_test,y_test_lr)
recall_score_test_lr = metrics.recall_score(y_test,y_test_lr)
precision_score_test_lr = metrics.precision_score(y_test,y_test_lr)

print(metrics.classification_report(y_test_lr, y_test))
```

```python
algo_holder("Logistic Regression", acc_test_lr, f1_score_test_lr, recall_score_test_lr, precision_score_test_lr)
```

<h1>6.2 k-Nearest Neighbors</h1>
<p>
A k-nearest neighbors (KNN) classifier is a type of machine learning algorithm used for classification tasks. It works by finding the k nearest data points to a new input data point and assigning the class based on the majority of the classes among these neighbors.

To understand how a KNN classifier works, imagine you want to classify a flower based on its features like petal length and width.

The KNN algorithm starts by looking at the k closest flowers to the new input flower based on their features. For example, if k = 5, the algorithm would find the 5 closest flowers to the new input flower.

Next, the algorithm looks at the classes of these k closest flowers and assigns the class to the new input flower based on the majority of the classes among these neighbors. For example, if 3 of the 5 closest flowers are classified as setosas and 2 are classified as versicolors, the algorithm would assign the new input flower the class of setosa.

The value of k determines how many neighbors are considered in the classification process. If k is too small, the algorithm may be sensitive to noise in the data and may not generalize well. If k is too large, the algorithm may not capture local variations in the data.

In summary, a KNN classifier is like a voting system that looks at the k closest data points to a new input data point and assigns the class based on the majority of the classes among these neighbors. It is a simple but effective algorithm for classification tasks.</p>

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=1)

knn.fit(X_train, y_train)
```

```python
y_train_knn = knn.predict(X_train)
y_test_knn = knn.predict(X_test)
```

```python
cm = metrics.confusion_matrix(y_train_knn, y_train)
cm_display = metrics.ConfusionMatrixDisplay(cm).plot()
```

```python
acc_test_knn = metrics.accuracy_score(y_test,y_test_knn)
f1_score_test_knn = metrics.f1_score(y_test,y_test_knn)
recall_score_test_knn = metrics.recall_score(y_test,y_test_knn)
precision_score_test_knn = metrics.precision_score(y_test,y_test_knn)

print(metrics.classification_report(y_test_knn, y_test))
```

```python
algo_holder("KNN", acc_test_knn, f1_score_test_knn, recall_score_test_knn, precision_score_test_knn)
```

<h1>6.3 Naive Bayes</h1>
<p>A Naive Bayes classifier is a type of machine learning algorithm used for classification tasks. It works by using probability theory to predict the most likely class for a given input.

To understand how a Naive Bayes classifier works, imagine you want to classify whether an email is spam or not based on its words.

The Naive Bayes algorithm starts by calculating the probability of each word appearing in a spam email and in a non-spam email based on a training dataset. For example, it might calculate that the word "money" appears in 80% of spam emails but only 10% of non-spam emails.

Next, the algorithm looks at the words in the input email and calculates the probability that the email is spam or not based on the probability of each word appearing in a spam email or a non-spam email.

For example, if the email contains the words "money" and "sale", the algorithm would calculate the probability that the email is spam given the appearance of these words in the email. It does this by multiplying the probability of "money" appearing in a spam email (0.8) by the probability of "sale" appearing in a spam email (let's say, 0.6), and then multiplying that by the overall probability of a spam email (let's say, 0.4). It does the same calculation for the non-spam probability.

Finally, the algorithm chooses the class with the highest probability as the predicted class for the input email. For example, if the spam probability is 0.12 and the non-spam probability is 0.01, the algorithm would predict that the email is spam.

In summary, a Naive Bayes classifier is like a probability calculator that calculates the probability of an input belonging to each class based on the probability of each feature appearing in that class. It then chooses the class with the highest probability as the predicted class for the input.</p>

```python
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()

nb.fit(X_train, y_train)
```

```python
y_train_nb = nb.predict(X_train)
y_test_nb = nb.predict(X_test)
```

<h1>6.4 Decision Tree</h1>
<p>A decision tree is a type of machine learning algorithm used for classification tasks. It works by creating a tree-like model of decisions and their possible consequences.

To understand how a decision tree classifier works, imagine you want to classify whether a fruit is an apple or an orange based on its features like color, diameter, and weight.

The decision tree algorithm starts by selecting the most important feature, which is the one that best separates the apples from the oranges. For example, if color is the most important feature, the algorithm might ask "Is the fruit red or orange?" If the answer is red, it might classify the fruit as an apple, and if the answer is orange, it might classify the fruit as an orange.

Next, the algorithm looks at the next most important feature and repeats the process. For example, it might ask "Is the diameter of the fruit greater than 8 cm?" If the answer is yes, it might classify the fruit as an apple, and if the answer is no, it might classify the fruit as an orange.

The algorithm continues this process, selecting the most important feature and splitting the data into smaller groups based on their values until all the data is classified. The result is a tree-like model of decisions and their possible consequences.

To classify a new fruit, the algorithm starts at the top of the tree and follows the path of decisions until it reaches a final classification.

In summary, a decision tree classifier is like a flowchart that asks a series of questions to classify data based on their features. It selects the most important feature and splits the data into smaller groups until all the data is classified.</p>

```python
from sklearn.tree import DecisionTreeClassifier


dt = DecisionTreeClassifier()

dt.fit(X_train, y_train)
```

```python
y_train_dt = dt.predict(X_train)
y_test_dt = dt.predict(X_test)
```

<h1>6.5 Random Forest</h1>
<p>
Random Forest is a popular machine learning algorithm used for classification tasks. It works by combining multiple decision trees to make a final prediction.

To understand how a random forest classifier works, imagine you want to classify whether an animal is a dog or a cat based on its features like size, weight, and color.

A decision tree is like a flowchart that asks a series of yes-or-no questions to classify the animal. For example, the first question might be "Is the animal's weight greater than 10 kg?" If the answer is yes, the decision tree might go down one branch, and if the answer is no, it might go down a different branch.

A random forest classifier takes this idea a step further by creating multiple decision trees and combining their predictions to make a final decision. Each decision tree is built using a different subset of the data and a random selection of features. This helps to reduce overfitting, which is when a model is too complex and performs well on the training data but poorly on new data.

Once all the decision trees are built, the random forest classifier makes a final prediction by combining the predictions of all the decision trees. For example, if 60% of the decision trees predict the animal is a dog and 40% predict it's a cat, the random forest classifier would classify the animal as a dog.

In summary, a random forest classifier is like a group of decision trees working together to make a final decision. Each tree makes its own prediction, and the final prediction is based on the collective results of all the trees.</p>

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=10)

rf.fit(X_train, y_train)
```

<h1>6.6 Gradient Boosting</h1>
<p>Gradient Boosting is a type of machine learning algorithm used for both regression and classification tasks. It works by combining several weak models to create a strong model that makes accurate predictions.

To understand how gradient boosting works, imagine you want to predict the price of a house based on its features like number of bedrooms, bathrooms, and square footage.

The gradient boosting algorithm starts by creating a simple model, such as a decision tree, to predict the house prices. However, this initial model may not be very accurate.

Next, the algorithm identifies the data points that the initial model predicted incorrectly and assigns them a higher weight. It then creates a second model, such as another decision tree, that focuses on predicting the incorrect predictions from the first model.

The algorithm then combines the predictions from the first and second models to make a more accurate prediction of the house prices. This process is repeated, with the algorithm building more and more models, each focusing on the incorrect predictions from the previous models.

Finally, the algorithm combines all of the models to make a final prediction of the house prices. This final model, created by combining the outputs of all the individual models, is typically much more accurate than the initial model.

In summary, gradient boosting is like building a team of models, where each model focuses on improving the accuracy of the previous models by identifying and correcting their mistakes. The end result is a strong model that makes accurate predictions.</p>

```python
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier(n_estimators=300, max_depth=5)

gb.fit(X_train, y_train)
```

```python
y_train_gb = gb.predict(X_train)
y_test_gb = gb.predict(X_test)
```

```python
cm = metrics.confusion_matrix(y_train, y_train_gb)
cm_display = metrics.ConfusionMatrixDisplay(cm).plot()
```

```python
acc_test_gb = metrics.accuracy_score(y_test,y_test_gb)
f1_score_test_gb = metrics.f1_score(y_test,y_test_gb)
recall_score_test_gb = metrics.recall_score(y_test,y_test_gb)
precision_score_test_gb = metrics.precision_score(y_test,y_test_gb)

print(metrics.classification_report(y_test, y_test_gb))
```

```python
algo_holder("Gradient Boosting", acc_test_gb, f1_score_test_gb, recall_score_test_gb, precision_score_test_gb)
```

<h1>6.7 Support Vector Machine</h1>
<p>Support Vector Machines, or SVM for short, is a type of machine learning algorithm used for classification tasks. It works by finding the best possible boundary between different classes in the input data.

To understand how SVM works, imagine you want to predict whether a fruit is an apple or an orange based on its color and size.

The SVM algorithm starts by plotting the input data points on a graph. Each data point represents a fruit and is placed in the graph based on its color and size.

Next, the algorithm identifies the best possible boundary, called the decision boundary, that separates the two classes of fruits. The decision boundary is found by looking for the widest possible gap between the two classes.

However, sometimes the data is not linearly separable, meaning the decision boundary cannot be a straight line. In that case, SVM uses a technique called kernel trick to transform the input data into a higher-dimensional space, where the decision boundary can be a hyperplane.

Once the decision boundary is found, the SVM algorithm assigns a label to the fruits on each side of the boundary. Any new fruit that falls on one side of the decision boundary will be classified as that class.

In summary, SVM is like finding the best possible line or curve that separates the input data into different classes. It's a powerful algorithm that works well even when the data is not easily separable, thanks to the kernel trick.</p>

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV


param_grid = {'kernel': ['rbf', 'linear', 'sigmoid'], 'gamma': [0.1, 0.2, 0.3]}
svc = GridSearchCV(SVC(), param_grid, cv=5)
svc.fit(X_train, y_train)
print("Best parameters: ", svc.best_params_)
print("Best score: ", svc.best_score_)
```

<p>GridSearchCV is a method for hyperparameter tuning in machine learning. It works by searching over a range of hyperparameters for a given algorithm to find the combination of hyperparameters that results in the best performance.

In the context of SVM, GridSearchCV can be used to find the best combination of hyperparameters for the SVC (Support Vector Classifier) algorithm. The hyperparameters for SVC include the regularization parameter C, the kernel type (linear, polynomial, RBF, sigmoid), and the kernel coefficient for some kernels.</p>

```python
y_test_svc = svc.predict(X_test)
y_train_svc = svc.predict(X_train)
```

```python
cm = metrics.confusion_matrix(y_train, y_train_svc)
cm_display = metrics.ConfusionMatrixDisplay(cm).plot()
```

```python
acc_test_svc = metrics.accuracy_score(y_test,y_test_svc)
f1_score_test_svc = metrics.f1_score(y_test,y_test_svc)
recall_score_test_svc = metrics.recall_score(y_test,y_test_svc)
precision_score_test_svc = metrics.precision_score(y_test,y_test_svc)

print(metrics.classification_report(y_test, y_test_svc))
```
